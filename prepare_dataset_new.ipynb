```python
start_time = datetime.datetime.now()
print('start :',start_time)
for device in devices:
    print("-------------------",device,"-------------------------------")
    out_base = f"/media/di/data/lee/nia/Former-DFER/nia/data/all_clips/{device}/"
    flist_all = glob(nas_dir+f"/*/processing/S1/???/T1/{device}/RGB/*.mp4")
    flist_all.sort()
    video_last = len(glob(f"/media/di/data/lee/nia/Former-DFER/nia/data/all_clips/{device}/v*/frame_15.png"))
    n_lists = len(glob("/media/di/data/lee/nia/Former-DFER/nia/data/clip_list*.txt"))
    os.makedirs(out_base,exist_ok=True)
    nframes=32
    with open(f"/media/di/data/lee/nia/Former-DFER/nia/data/{device}_clip_list{n_lists:02d}.txt", "w") as f:
        for fn_video in flist_all:
            print(fn_video)
            state = utils.get_state(fn_video)
            dir_out = out_base + f"v{video_last:06d}/"
            print(dir_out)
            if not os.path.isdir(dir_out):   #// This fails
                os.makedirs(dir_out)        
            #try:
            if True:
                nframes_done = utils.crop_frames(fn_video, dir_out, 
                            nframes=nframes, 
                            sampling_freq = 8, 
                            #rot=cv2.ROTATE_90_COUNTERCLOCKWISE,
                            factor=2.0)
                video_last +=1
                if nframes_done < nframes:
                    continue            
                f.write(f"{dir_out} {nframes_done} {state}\n")
                print("done", video_last, "state", state)
            #except:
            else:
                continue
    f.close()
end_time = datetime.datetime.now()
print('end :',end_time)
```
