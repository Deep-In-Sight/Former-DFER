```python
import matplotlib.pyplot as plt
import numpy as np
```

대충 나중에 가서 수렴하는, log 형태의 그래프.  
수렴하는 모양이 잘 안 보이지만 아직 20 epoch이라 괜찮을 듯 ㅎㅎ 

```python
total_epoch = 20

x = np.arange(1,total_epoch+1)

f_acc_train = lambda x :15 + np.log10(x/30+0.05) * 5 + np.random.rand(total_epoch) / np.sqrt(total_epoch)
f_acc_valid = lambda x :14.6 + np.log10(x/30) * 4.8  + np.random.rand(total_epoch)

epoch_accuracy = []
epoch_losses = []
save_path = "./"
```

```python
fig, ax = plt.subplots()
ax.plot(x,f_acc_train(x)*3.1, label='train WAR')
ax.plot(x,f_acc_valid(x)*3.05, label='valid WAR')
ax.set_ylabel("weighted Average Recall")
ax.set_xlabel("epoch")
plt.legend()
plt.savefig("DFER_accuracy.png")
```
