```python
import shutil
```

```python
fin = open("./annotation/set_1.txt", "r")
lines = fin.read().split("\n")
lines.pop()    
with open("./annotation/set_1_cleaned.txt", "w") as f:
    for i, ll in enumerate(lines):
        if int(ll.split(" ")[1]) >= 16:
            f.write(ll+"\n")
        else:
            # remove dir
            try:
                shutil.rmtree(ll.split(" ")[0])
            except:
                continue
        print(i)
```
